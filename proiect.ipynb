```python
# Eliminam coloana label pentru a ramane doar cu caracterisiticile pozelor
features = df.drop('L ', axis=1)
label = df['L ']

#convertim tipul de date din panda in ndarray 
testFeatures =df_t.to_numpy() 

#Extragem dimensiunea matricei de caracteristici
n,m = features.shape 
print(m)
print(n)

#Construiti si rezolvati problema cmmp utilizand comanda lstqr


# Descompunerea QR comanda in Python
q,r = np.linalg.qr(np.copy(A))
print('--------q-----\n', q)
print('--------r-----\n',r)

def  TORT (A):
    m, n  = np.shape(A)
    U = np.zeros((m,n))
    p = min(m-1,n)
    beta = np.zeros((p,1))
    for k in range(p):
        # Calcularea lui sigma
        s=0
        for i in range(k,m):
            s = s+ A[i, k]**2
        sigma = np.sign(A[k,k])*np.sqrt(s)
        
        if sigma != 0:
            # Calcularea vectorilor householder
            U[k,k] = A[k,k] +sigma
            for i in range(k+1,m):
                U[i,k] = A[i,k]            
            beta[k] = sigma*U[k,k]
            #Aplicarea reflectorului Householder pe A
            for j in range(k,n):
                tau = 0
                for q in range(k,m):
                    tau = tau + U[q,k]*A[q,j]
                tau =tau/beta[k]
                for i in range(k,m):
                    A[i,j] = A[i, j] - tau *U[i,k]
                    
    return U, A, beta

# Algoritmul UTRIS
def Utris(U, label):
    n = len(U)
    x = np.zeros((n,1))
    for i in range(n-1,-1, -1):
        s = label[i]
        for j in range(i+1,n):
            s = s -U[i][j]*x[j]
        x[i] = s/U[i][i]
    return x


#Triangularizarea ortogonala a lui A
U,R,beta  = TORT(A)
print('----------U------------\n', U)
print('-----------R-----------\n',R)

# Aplicarea Reflectorilor asupra lui b
for k in range(n):
    tau=0
    for i in range(k,m-1):
        tau = tau + U[i,k]*label[i]
    tau = tau/beta[k]
    for i in range(k,m-1):
        label[i] = label[i] - tau*U[i,k]

#Calcularea solutiei CMMP
x  =Utris(R[0:n,:], label[0:n,0])

print('-----------x------------\n',x)

#A = np.append(features, np.ones((n,1)), axis = 1) 
#x = np.linalg.lstsq(A, label, rcond = None)[0]
print(x)

```
